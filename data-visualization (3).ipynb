1. Heat Map

```python
df.corr().style.format("{:.3}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

2. Boxplot

```python
fig,ax=plt.subplots(figsize=(18,12))
sns.boxplot(x="pcc",y="age",hue="ba",data=df)
```

3. Countplot

```python
sns.countplot(x='ba',data=df,hue='pcc',palette='BuPu',linewidth=3)
```

4. Barplot

```python
sns.barplot(x="bp", y="pcc", hue="ba", data=df,palette="PuBu")
```

5. Scatterplot

```python
sns.scatterplot(x="age", y="bp",hue="ba",data=df,palette="cool")
```

6. LM plot

```python
sns.lmplot(x="bp", y="al", col="pcc", hue="ba",
           data=df)
```

7. Point plot

```python
sns.pointplot(x = "bp",y = 'al',data = df,color = "green",alpha = .6)
```

8. Pairplot

```python
sns.pairplot(df,hue="ba",vars=["age","bp"],palette="husl")
```

 9. Joint plot

```python
sns.set(style="white")
sns.jointplot(x="sg",y="age",kind="kde",color="r",data=df)
```

10. Catplot

```python
sns.catplot(x="sg", y="age",hue="ba", col="pcc",data=df, kind="swarm",height=5.7, aspect=.5)
```

11. Implot

```python
sns.lmplot(x="sg", y="age", col="ba", hue="pcc", data=df)
```

12. Lineplot

```python
sns.lineplot(x="sg", y="age", hue="ba", data = df)
```

13. Relplot

```python
sns.relplot(
    data=df,
    x="age", y="bp",
    hue="pcc", size="sg", col="ba",
    kind="line", size_order=["T1", "T2"],
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

14. Violinplot

```python
sns.violinplot(data=df, x="age", y="bp", hue="al")
```

15. Stripplot

```python
sns.stripplot(x="age", y="bp",
              data=df, dodge=True, alpha=.25, zorder=1)
```
